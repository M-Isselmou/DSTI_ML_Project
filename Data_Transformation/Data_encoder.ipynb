```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

cyberdata = pd.read_csv("cyberdata_clean.csv")
cyberdata
```

```python
from sklearn.preprocessing import LabelEncoder

# Colonnes à encoder (exclure les adresses IP)
columns_to_encode = [
    "Protocol", "Packet Type", "Traffic Type", "Attack Type", "Attack Signature",
    "Action Taken", "Severity Level", "Network Segment", "Log Source", "City",
    "Region", "Browser", "Operating System", "Anomaly Score Category", "Packet Length Category"
]

# Appliquer le Label Encoding
label_encoder = LabelEncoder()
cyberdata_label_encoded = cyberdata.copy()  # Copie du DataFrame original

for col in columns_to_encode:
    cyberdata_label_encoded[col] = label_encoder.fit_transform(cyberdata[col])

# Afficher les premières lignes du DataFrame encodé
print(cyberdata_label_encoded.head())
```

```python
csv_file_path = "cyberdata_ip_city.csv"
cyberdata_label_encoded.to_csv(csv_file_path, index=False)
```
